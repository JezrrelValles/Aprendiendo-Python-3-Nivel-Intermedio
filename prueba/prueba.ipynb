```python
import requests
from bs4 import BeautifulSoup
```

```python
url = requests.get('https://juliocepeda.com/ni-os.html')

soup = BeautifulSoup(url.content, "html.parser")
```

```python
soup
```

```python
precios = []

precio_items = soup.find_all('span', class_="price")

for item in precio_items:
     precio = item.text
     precios.append(precio)
```
